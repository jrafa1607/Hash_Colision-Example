```python
#Exemplo de Colisão de Hashes
#Import das Bibliotecas

import os
import cv2
import hashlib
import matplotlib as mpl
from matplotlib import pyplot as plot
```

```python
#Importando a imagem

img1 = cv2.imread('D:/OneDrive/OneDrive - Etec Centro Paula Souza/Databases/Images/md5_file1.jpg')
img2 = cv2.imread('D:/OneDrive/OneDrive - Etec Centro Paula Souza/Databases/Images/md5_file2.jpg')
```

```python
#Visualizando a Imagem 01
plot.imshow(img1,cmap='gray')
plot.show()
```

```python
#Visualizando a Imagem 02
plot.imshow(img2,cmap='gray')
plot.show()
```

```python
#Na condução dos experimentos, percebeu-se que se calcular o Hash com o input da Imagem pelo CV. O valor estará errado!
#Isso porque o Python ajusta o import da imagem. Desta forma, é necessário ler a imagem informando que é um arquivo binário.
```

```python
#Para Exemplificar, será calculado primeiro o Hash das imagens importadas pelo CV
#Os dois valores serão diferentes, e não será possível evidenciar a colisão.

print(hashlib.md5(img1).hexdigest())
print(hashlib.md5(img2).hexdigest())
```

```python
#Agora, será feito a leitura das imagens, informando ao Python que deseja-se manter a estrutura binária da imagem
file1 = 'D:/OneDrive/OneDrive - Etec Centro Paula Souza/Databases/Images/md5_file1.jpg'
file2 = 'D:/OneDrive/OneDrive - Etec Centro Paula Souza/Databases/Images/md5_file2.jpg'

with open(file1, 'rb') as f1:
    data1 = f1.read()
with open(file2, 'rb') as f2:
    data2 = f2.read()
```

```python
#Print dos Resultados
print("O valor Hash da Imagem 01 é:", hashlib.md5(data1).hexdigest())
print("O valor Hash da Imagem 02 é:", hashlib.md5(data2).hexdigest())
```

```python
#Exemplo de Colisão de Hash Sha1 (Com dois arquivos .pdf)
file1 = 'D:/OneDrive/OneDrive - Etec Centro Paula Souza/Databases/Documents/shattered-1.pdf'
file2 = 'D:/OneDrive/OneDrive - Etec Centro Paula Souza/Databases/Documents/shattered-2.pdf'
```

```python
#Leitura dos Arquivos Pdf
with open(file1, 'rb') as f1:
    data1 = f1.read()
with open(file2, 'rb') as f2:
    data2 = f2.read()
```

```python
#Exibição dos arquivos pdfs
os.startfile(file1)
os.startfile(file2)
```

```python
#Print dos Resultados
print("O valor Hash do pdf 01 é:", hashlib.sha1(data1).hexdigest())
print("O valor Hash do pdf 02 é:", hashlib.sha1(data2).hexdigest())
```
