```python
#Exemplo de Colisão de Hashes
#Import das Bibliotecas

import os
import cv2
import hashlib
import matplotlib as mpl
from matplotlib import pyplot as plot
```

```python
#Exemplo de Colisão com MD5
#Importando as imagens "diferentes"

img1 = cv2.imread('D:/Google Drive/Github/HashFunctions/Images/md5_file1.jpg')
img2 = cv2.imread('D:/Google Drive/Github/HashFunctions/Images/md5_file2.jpg')
```

```python
#Visualização das imagens 

plot.imshow(img1,cmap='gray');
plot.show()
plot.imshow(img2,cmap='gray');
plot.show()
```

```python
#Cáculo do hash MD5 das duas imagens:
print(hashlib.md5(img1).hexdigest())
print(hashlib.md5(img2).hexdigest())
```

```python
#Na fase anterior, calculou-se o hash das imagens na memória do computador carrwgadas pela biblioteca openCV
#O próximo passo é carregar as imagens informando que é um arquivo binário.
```

```python
#Import das imagens, mantendo a estrutura binária
file1 = 'D:/Google Drive/Github/HashFunctions/Images/md5_file1.jpg'
file2 = 'D:/Google Drive/Github/HashFunctions/Images/md5_file2.jpg'

with open(file1, 'rb') as f1:
    data1 = f1.read()
with open(file2, 'rb') as f2:
    data2 = f2.read()
```

```python
#Cáculo do hash MD5 das duas imagens:
print("O valor Hash da Imagem 01 é:", hashlib.md5(data1).hexdigest())
print("O valor Hash da Imagem 02 é:", hashlib.md5(data2).hexdigest())
```

```python
#Colisão de Hash, pois mesmo com imagens "diferentes", o valor de Hash foi o mesmo.
#Documentation: https://freemanlaw.com/hash-collisions-explained/
```

```python
#Outro Exemplo de Colisão, desta vez com o algoritmo Sha1 e arquivos Pdf
file1 = 'D:/Google Drive/Github/HashFunctions/Documents/shattered-1.pdf'
file2 = 'D:/Google Drive/Github/HashFunctions/Documents/shattered-2.pdf'
```

```python
#Leitura dos Arquivos Pdf
with open(file1, 'rb') as f1:
    data1 = f1.read()
with open(file2, 'rb') as f2:
    data2 = f2.read()
```

```python
#Exibição dos arquivos pdfs
os.startfile(file1)
os.startfile(file2)
```

```python
#Print dos Resultados
print("O valor Hash do pdf 01 é:", hashlib.sha1(data1).hexdigest())
print("O valor Hash do pdf 02 é:", hashlib.sha1(data2).hexdigest())
```
