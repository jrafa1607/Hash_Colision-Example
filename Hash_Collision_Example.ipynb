```python
#import Libs
import os
import cv2
import hashlib
import numpy as np
import urllib.parse
import urllib.request
import matplotlib as mpl
from matplotlib import pyplot as plot
```

```python
#Function to Load a Image from Github
def display_image_from_url(image_url):
    with urllib.request.urlopen(image_url) as url:
        s = url.read()
    img = cv2.imdecode(np.frombuffer(s, np.uint8), -1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img, s
```

```python
#Function to Calc Hash
def calculate_and_display_hashes(file_input):
  data = file_input

  print(f"\n--- Hash Values for: {file_input if isinstance(file_input, str) else 'binary data'} ---")
  print("MD5:", hashlib.md5(data).hexdigest())
  print("SHA1:", hashlib.sha1(data).hexdigest())
  print("SHA256:", hashlib.sha256(data).hexdigest())
  print("SHA512:", hashlib.sha512(data).hexdigest())
```

```python
#Images URL's from Github
image_url_1 = 'https://raw.githubusercontent.com/jrafa1607/Hash_Colision-Example/main/Dataset/img01.jpg'
image_url_2 = 'https://raw.githubusercontent.com/jrafa1607/Hash_Colision-Example/main/Dataset/img02.jpg'
```

```python
#Load the Images
img1, data1 = display_image_from_url(image_url_1)
img2, data2 = display_image_from_url(image_url_2)
```

```python
#Show the Images
fig, axes = plot.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img1)
axes[0].set_title('Image 1')
axes[0].axis('off')
axes[1].imshow(img2)
axes[1].set_title('Image 2')
axes[1].axis('off')
plot.tight_layout()
plot.show()
```

```python
#Show the Hash Results ( MD5 Collision )
calculate_and_display_hashes(data1)
calculate_and_display_hashes(data2)
```
